```python
df=pd.read_csv("Mobile_Dataset.csv")
df.head(7)
```

What is the range of the smartphones price?

```python
print(df['Price(Rs.)'].max()-df['Price(Rs.)'].min())
```

What is the highest and lowest Specs_score of the smartphones?

```python
print("The Highest Specs_score is ",df['Specs_Score'].max())
print("The lowest Specs_score is ",df['Specs_Score'].min())
```

-------------------------------------------------------------------------------------------------------------------------------

What is the spread in the smartphones price?

```python
print("The variation of smartphone's price is" ,df['Price(Rs.)'].var())
print("The standard deviation of smartphone's price is", df['Price(Rs.)'].std())
```

Here we can see that spread in smartphone's price can be visualised by standard deviation because the standard deviation provides a more interpretable measure of the average deviation of individual prices from the mean. It indicates that, on average, smartphone prices deviate by approximately 31922.41 Rs. from the mean price.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

what is the best representation for model name and specs score?

```python
plt.figure(figsize=(10,11))
plt.bar(df['Name'],df['Specs_Score'],label="Specs_score vs Mobile Name")
plt.xlabel('name')
plt.ylabel('specs_score')
plt.xticks (rotation='vertical')
plt.legend()
plt.show()
```

How to measure the distribution or spread of data points using price and main camera?

```python
plt.figure(figsize=(10,9))
plt.scatter(df['Price(Rs.)'],df['Main_Camera'],marker='o',color='b')
plt.ylabel('Main_Camera')
plt.xlabel('Price(Rs.)')
plt.xticks (rotation='vertical')
plt.grid(True)
plt.legend()
plt.show()
```

scatter plot spots data points that deviates from general data points help in finding the errors or unusal changes

Given Mobile_dataset.csv,create a bar graph to visually depict the price variation of Mobiles among different manufacturers.

```python
import pandas as pd
import matplotlib.pyplot as plt
df3=pd.read_csv("Mobile_dataset.csv")
plt.bar(df3['Name'],df3['Price(Rs.)'],label="Price vs Manufacturer",linewidth=0.5,width=0.6)
plt.xlabel('Manufacturers')
plt.ylabel('Price')
plt.xticks( rotation='vertical' )
plt.legend()
plt.show()
```

This barplot is to visualize different prices and their manufacturers

 Given Mobile_dataset.csv. plot a bar graph to visually represent the screen sizes of mobiles across manufacturers.

```python
import pandas as pd
import matplotlib.pyplot as plt
df4=pd.read_csv("Mobile_dataset.csv")
plt.bar(df4['display'],df4['Price(Rs.)'],label="Screen size vs Manufacturer",linewidth=0.7,width=0.3)
plt.xlabel('Display')
plt.ylabel('Price')
plt.xticks( rotation='vertical' )
plt.legend()
plt.show()
```

The barplot shows the relationship between screen size (display) and price for a particular manufacturer. There is a noticeable increase in price as the screen size increases, with an exception at 6.50 where there is a significant drop in price before it starts increasing again.
